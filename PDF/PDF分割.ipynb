```
#@title 引用需要的工具
import PyPDF2

import pprint

from google.colab import files

import pdfkit 
```

```
#@title 上傳PDF檔(目前只能上傳一個)
f=files.upload()
for fn in f.keys():
  print('User uploaded file "{name}" with length {length} bytes'.format(name=fn, length=len(f[fn])))
  File = open(fn,'rb')
```

```
#@title 顯示內部資料(只有英文能正常出現)
PDF = PyPDF2.PdfFileReader(File)
for fn in f.keys():
  print(fn)
print("共計："+str(PDF.numPages)+"頁")
p=1
print("_________________________________________________________________________\n")
for page in PDF.pages:
    print(page.extractText()+"\n")
    print("\t\tP"+str(p)+"\n\n")
    p+=1
    
print("\n_________________________________________________________________________")
```

```
#@title 輸入要切割的頁數，要結束時請打上end
variable_name = 1
out = PyPDF2.PdfFileWriter()
l=list()
#tmp=input()
#tmp=tmp.split(" ")
while 1:
  inp=input()
  if inp=="end":
    break
  else:
    out.addPage(PDF.getPage(int(inp)))

     
#pdfkit.from_string('','out.pdf')
out.write(open('out.pdf', 'wb'))

files.download("out.pdf")
!rm *.pdf
```
